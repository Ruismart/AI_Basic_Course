# 1. 从fundemantals.csv开始！

fundemantals.csv 是这些股票的年报数据

请用数据回答以下问题：

1. S&P500股票在2015年`net income`的均值是多少？最大值比最小值多多少？（每问10分，共计20分）
2. S&P500股票在2016年的固定资产（fixed assets）占总资产(total assets)比例的均值是多少？固定资产占总资产比例最小的股票是的代码（ticker symbol）是什么？（每问10分，共计20分）


```python
fdt = pd.read_csv('../nyse/fundamentals.csv')
pd.set_option('display.max_columns',None)      # set to display all columns
fdt.head()
```

```python
fdt1 = fdt[fdt['For Year']==2015.0]        # choose data of year 2015
fdt1_2015 = fdt1.filter(items=['For Year','Net Income']).groupby('For Year')   # choose required columns and group by year
fdt1_2015.mean()       # S&P500 2015年的均值
```

```python
fdt1_2015.max()     # S&P500 2015年的最大值
```

```python
fdt1_2015.min()    # S&P500 2015年的最小值
```

```python
fdt1_2015.max()-fdt1_2015.min()   # 最大值比最小值大多少
```

```python
fdt2 = fdt[fdt['For Year']==2016.0]         # choose data of year 2016
fdt2 = fdt2.filter(items=['For Year','Ticker Symbol','Fixed Assets','Total Assets'])  # choose required columns
fdt2['Ratio'] = fdt2['Fixed Assets'] / fdt2['Total Assets']   # add a new column which is required ratio 
fdt2.head()
```

```python
# S&P500股票在2016年的固定资产（fixed assets）占总资产(total assets)比例的均值
fdt2.filter(items=['For Year','Ratio']).groupby('For Year').mean()  
```

```python
# 比例最小的股票代码
low = fdt2[fdt2['Ratio']==fdt2['Ratio'].min()]   
low.filter(items=['Ticker Symbol','Ratio'])
```

# 2. 加入securities.csv~

securities.csv包含了这些股票的基本信息

1. 请列举出各个sector中的加入时间最早的股票名称（10分）
2. 请列举出每一个州中加入时间最晚的股票名称（10分）


```python
sec = pd.read_csv('../nyse/securities.csv')
sec.head()
```

```python
#
sec1 = sec.filter(items=['GICS Sector','Ticker symbol','Date first added'])  # choose required columns
sec1 = sec1.dropna(axis=0, how='any')    # filter out any rows with a NA value
sec1.head()
```

```python
#  group by 'GICS Sector' and each sector choose the earliest date
#        use .apply() to run a function with the data of each group
early = sec1.groupby(['GICS Sector']).apply(lambda dat: dat[dat['Date first added']==dat['Date first added'].min()] )
early  # 每个 Sector 中，加入最早的股票名称
```

```python
sec2 = sec.filter(items=['Address of Headquarters','Ticker symbol','Date first added'])  # choose required columns
sec2 = sec2.dropna(axis=0, how='any')          #  filter out any rows with a NA value
sec2.head()
```

```python
# extract 'State' from the last word of each address in 'Address of Headquarters'
sec2['State'] = [state.split(',')[-1] for state in sec2['Address of Headquarters']]
sec2.head()
```

```python
#  group by 'State' and each group choose the latest date
late = sec2.groupby(['State']).apply(lambda dat: dat[dat['Date first added']==dat['Date first added'].max()] )
pd.set_option('display.max_rows',None)  # display all rows
late  # 每个州加入最早的股票名称
```

# 3. merge!

现在你需要同时处理来自两个表中的信息了

1. 请思考，合并两个表的信息的时候，我们应该用什么样的准则对其它们（10分）
2. 请列举每个sector在2013-2016年累计Research&Development的总投入（10分）
3. 请列举出每个sector中，在2013-2016年累计Research&development投入最大的3家公司的名称以及投入的数值（20分）

1. 合并两个表信息的时候，要去看它们各自的行列数据有没有能对应上的  
   这两个表都有一列类似的Ticker Symbol,用它就可以把两个表的数据联系起来，需要注意数目和顺序可能并不一样

2,3 的思路，先从第一个表里得到每只股票在 2013-2016 的研发总投入，再从第二个表得到每只股票的对应sector 信息  
然后使用对应列合并两个数据，得到新的数据表格“sector 股票 研发投入”，最后用 sector 分组后作统计

```python
fdt3 = fdt[(fdt['For Year'] >= 2013.0) & (fdt['For Year'] <= 2016.0)]  # choose data of year 2013-2016
fdt3 = fdt3.filter(items=['Ticker Symbol','Research and Development'])   
fdt3 = fdt3.groupby('Ticker Symbol').sum()
fdt3.head()
```

```python
fdt4 = fdt3
fdt4['Ticker symbol'] = fdt3.index    # use 'symbol' instead of 'Symbol' to match two datasets
fdt4.head()
```

```python
sec3 = sec.filter(items=['GICS Sector','Ticker symbol'])
sec3.head()
```

```python
datam = pd.merge(sec3,fdt4)    # automatically merge two datasets using the column with the same column name
datam.head()
```

```python
datam1 = datam.filter(items=['GICS Sector','Research and Development'])
datam1 = datam1.groupby('GICS Sector').sum()
datam1  # 每个sector在2013-2016年累计Research&Development的总投入
```

```python
# 每个sector中，在2013-2016年累计Research&development投入最大的3家公司
datam2 = datam.groupby('GICS Sector').apply(lambda dat: dat.nlargest(3,'Research and Development'))
datam2 = datam2.filter(items=['Ticker symbol','Research and Development'])
datam2    # 有几个 sector 没有研发投入，如果要去掉可以先选出研发不为0的行
```

# 4. 现在让我们来看看更加复杂的数据

请导入price.csv，然后结合你的聪明才智回答以下问题（附加题，40分）

假设你是某基金公司的老板，现在对于每只股票，你都专门安排了一位负责它的交易员。公司规定每一位交易员手中的资金要么全部买入要么全部卖出（空仓，转化为现金）。假设2016年每一位交易员手中都有10000美元，假设他们都能够看到2016年全年的数据，假设他们都能抓住每一次机会，那么请问2016年底时，赚钱最多的股票是哪一只，赚了多少钱？

```python
price = pd.read_csv('../nyse/prices.csv')
price['ret'] = price['high']/price['low']
price.head()
```

思路：股票赚钱的方法是低买高卖，抓住每一次机会意味着总是在每天的价格最低点买入，最高点卖出   
                            （当然这只是最理想情况，实际上有可能最低点在最高点后面出现，不可能顺序买入卖出）  
                            （而且也不可能精确定位那两个时间点）  
                            假设的条件下，收益率 ret = high / low

```python
# test
test = price.head()
```

```python
[r for r in test['ret']]  # test 的收益率
```

```python
from functools import reduce
reduce(lambda x,y: x*y,[r for r in test['ret']])  # test 的总收益率， 总收益即是乘以 10000
```

```python
# apply 到 price 的每个 group
from functools import reduce
Ret = price.filter(items=['symbol','ret']).groupby('symbol').apply(lambda z: 10000 * reduce(lambda x,y: x*y,[r for r in z['ret']]))
Ret.head()
```

```python
Ret[Ret==Ret.max()]   # 最赚钱的股票和赚的钱
```
